```python
from vae.cond_conv_vae import CVAEConv2D
import torch
import pandas as pd
import numpy as np
from vae.utils import *
```

```python
set_seeds(0)
```

```python
model_dict = torch.load("test/condconv2d_vae.pt")
config = model_dict["model_config"]
model = CVAEConv2D(config)
model.load_weights(dict_to_load=model_dict)
```

```python
data = np.load("data/vol_surface_with_ret.npz")["surface"]
```

```python
surface = model.get_surface_given_conditions(c=torch.Tensor(data[10].reshape(1, 5, 5))).detach().cpu().numpy().reshape((5,5))
print(data[11])
print(surface)
```

```python
plot_surface_separate(data[11], surface)
```
