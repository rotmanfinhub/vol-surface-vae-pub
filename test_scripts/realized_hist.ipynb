```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
```

```python
data = np.load("data/vol_surface_with_ret.npz")
vol_surf_data = data["surface"]
ret_data = data["ret"]
skew_data = data["skews"]
slope_data = data["slopes"]
```

```python
print(vol_surf_data.shape)
print(skew_data.shape)
```

```python
x_ticker_formatter = ScalarFormatter(useMathText=True)
x_ticker_formatter.set_powerlimits((0, 0))
fig, ax = plt.subplots(1, 3, figsize=(18, 6))
sns.histplot(vol_surf_data[:, 3, 2], ax=ax[0])
ax[0].xaxis.set_major_formatter(x_ticker_formatter)
ax[0].set_xlabel("iv")
ax[0].set_ylabel("Frequency")
ax[0].set_title("IV[ttm=1y, K/S=1] distribution")

sns.histplot(skew_data, ax=ax[1])
ax[1].xaxis.set_major_formatter(x_ticker_formatter)
ax[1].set_xlabel("skew")
ax[1].set_ylabel("Frequency")
ax[1].set_title("skew distribution")

sns.histplot(slope_data[np.abs(slope_data) <= 2], ax=ax[2])
ax[2].xaxis.set_major_formatter(x_ticker_formatter)
ax[2].set_xlabel("slope")
ax[2].set_ylabel("Frequency")
ax[2].set_title("slope distribution")
plt.tight_layout()
plt.show()
```
